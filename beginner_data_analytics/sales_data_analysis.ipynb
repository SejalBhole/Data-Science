```python
import pandas as pd
```

```python
data = {
    "Product": ["Laptop", "Mouse", "Chair", "Laptop", "Mouse"],
    "Category": ["Electronics", "Electronics", "Furniture", "Electronics", "Electronics"],
    "Quantity": [1, 2, 1, 1, 3],
    "Price": [50000, 500, 3000, 50000, 500]
}

df = pd.DataFrame(data)

df
```

```python
#Create Revenue column
df["Revenue"] = df["Quantity"] * df["Price"]
df
```

```python
df["Revenue"].sum()
```

```python
#top selling product
df.groupby("Product")["Quantity"].sum()
```

```python
#product that gives highest revenue
df.groupby("Product")["Revenue"].sum()
```

```python
#category that gives highest revenue
df.groupby("Category")["Revenue"].sum()
```

```python
#category that solds most items
df.groupby("Category")["Quantity"].sum()
```
