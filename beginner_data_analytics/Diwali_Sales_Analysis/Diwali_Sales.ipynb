Here, **drop()** - Used to remove specific rows or columns by name or index.

You manually tell what to remove.

**axis=1 →** apply the operation on columns (not rows).
**inplace=True →** make changes in the same original dataframe, don’t create a new one.

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
#to delete null values of amount column
df.dropna(inplace=True)
```

Used to remove rows or columns that contain missing values (NaN).

**Exploratory Data Analysis**

**For Gender**

```python
ax = sns.countplot(x='Gender' , data=df)
```

```python
ax = sns.countplot(x='Gender' , data=df)

#Loops through each group of bars in the chart
#Adds numbers (labels) on top of each bar
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Gender', y='Amount', data=sales_gen)
```

**For State**

```python
#total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sales_state
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


#without sns.set
sns.barplot(x='State', y='Orders', data=sales_state)
```

**sns.set() →** Sets the default style/settings for seaborn plots.

**🔹 rc={'figure.figsize': (15, 5)} →** rc means Runtime Configuration.

It is used to change default settings of plots like:

figure size

font size

colors

grid style

**figure.figsize →** Sets the size of the graph.

**(15, 5) →** Width = 15 inches, Height = 5 inches.

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


#with sns.set
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
#total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y='Amount', data=sales_state)

sales_state
```

from the above graph we can see that most of the orders are from uttar pradesh, maharashtra and kernataka

For Marital Status
